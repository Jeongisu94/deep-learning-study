```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid(x):
    
    # np.exp(x) 는 자연상수(e)에 지수가 x인 수식
    
    return 1 / (1 + np.exp(-x))
```

```python
# -5.0부터 5.0 전까지 0.1간격의 넘파이 배열 x 생성
# 즉, [-5.0, -4.9, ....... , 4.8, 4.9]
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
# x가 배열이므로 sigmoid()의 결과도 배열
```

```python
plt.plot(x, y)
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.title("<< sigmoid function >>")
plt.show()
```
