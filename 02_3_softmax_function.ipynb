```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def softmax(x):
    exp_x = np.exp(x)
    sum_exp_x = np.sum(exp_x)
    y = (exp_x / sum_exp_x)
    
    return y
```

```python
# 소프트맥스 함수 구현 시 overflow 문제를 주의해야한다.
# 소프트맥스는 지수 함수함수라서 쉽게 컴퓨터의 데이터 표현범위를 넘을 수 있다.
# 소프트맥스 함수의 지수에 임의의 정수 c 를 더하거나 빼도 결과는 바뀌지 않는다는 점을 이용해서 오버플로를 막는다. 
# 자세한 설명은 블로그(junsoofeb.github.io) 참고

def softmax_safe(x):
    c = np.max(x) # 입력 신호 중 최댓값
    exp_x = np.exp(x - c) # 지수에 c를 빼서 오버플로 방지
    sum_exp_x = np.sum(exp_x)
    y = (exp_x / sum_exp_x)
    
    return y
```

```python
x = np.arange(-5.0, 5.1, 2.0)
y = softmax(x)
# x가 배열이므로 softmax()의 결과도 배열

# 소프트맥스 함수의 출력의 총합은 1
# 확률로 해석할 수 있음.
```

```python
plt.plot(x, y)
plt.xlabel("x") # x축 이름
plt.ylabel("y (probability)") # y축 이름
plt.ylim(-0.1, 1.1) # y축 범위 지정
for a,b in zip(x, y): 
    plt.text(a, b, str(b))
plt.title("<< softmax function >>")
plt.show()
```

```python
# overflow 주의한 버전

x1 = np.arange(-5.0, 5.1, 2.0)
y1 = softmax_safe(x1)
```

```python
plt.plot(x1, y1)
plt.xlabel("x") # x축 이름
plt.ylabel("y (probability)") # y축 이름
plt.ylim(-0.1, 1.1) # y축 범위 지정
for a,b in zip(x, y): 
    plt.text(a, b, str(b))
plt.title("<< softmax function(safe) >>")
plt.show()
```
